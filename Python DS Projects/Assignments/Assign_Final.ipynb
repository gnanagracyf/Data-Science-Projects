```python
def date_sorter():
    import re
    # Your code here
    #df_new = df.to_frame()
    df_new = df.rename(columns={0:"content"},inplace=True)
    
    

    re1 = '[0]?[0-9]?[0-9][/-][0-9]?[0-9][/-][19|20]?[0-9]{2,}'
    re2 = '(?:\d{1,2} )?(?:Jan|Feb|Mar|Apr|May|Jun|Jul|Aug|Sep|Oct|Nov|Dec)[a-z]*[.]?[-\s,.]?(?:\d{1,2})?[a-z]{,2}[,]?[-\s]?\d{4}'
    re3 = '[\D|\s][0-9]?[0-9][/][19|20]\d{2,}'
    re5 = '^[0-9]?[0-9][/][19|20]\d{2,}'
    re4 = '[\D|\s]\d{4}[\D|\s]'
    re6 = '^[19|20]\d{2,}'
    df_new['date'] = df_new['content'].apply(lambda x: re.compile("(%s|%s|%s|%s|%s|%s)" % (re1, re2, re3, re4, re5, re6)).findall(x))
    
    new_series = df_new['date'].apply(lambda x: ', '.join(x))
    new_series = new_series.apply(lambda x : x.strip())
    new_series = new_series.apply(lambda x: re.sub("9/36/308","",x))
    new_series = new_series.apply(lambda x: re.sub("/7787 ,","",x))
    new_series = new_series.apply(lambda x : re.sub(r"^[~(.//]","", x))
    new_series = new_series.apply(lambda x : re.sub(r"\D$","",x))
    new_series = new_series.apply(lambda x : re.sub(r"-","/",x))
    new_series = new_series.apply(lambda x: re.sub("r1978","1978",x))
    new_series = new_series.apply(lambda x: re.sub("y1974","1974",x))
    new_series = new_series.apply(lambda x: re.sub("y1983","1983",x))
    new_series = new_series.apply(lambda x: re.sub("s1981","1981",x))
    new_series = new_series.apply(lambda x: re.sub("n4/2004","4/2004",x))
    new_series = new_series.apply(lambda x: re.sub("e12/2007","12/2007",x))
    new_series = new_series.apply(lambda x: re.sub("011/14/83","11/14/83",x))
    new_series = new_series.apply(lambda x: re.sub("/7787 ,","",x))
    new_series = new_series.apply(lambda x: re.sub("Janaury ","Jan ",x))
    new_series = new_series.apply(lambda x: re.sub("Decemeber ","Dec ",x))
    
    date_series = pd.to_datetime(new_series)
    res = date_series.sort_values().index
    
    #return pd.Series(res)
    return df_new
```
