```python
%matplotlib notebook
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

np.random.seed(12345)

df = pd.DataFrame([np.random.normal(32000,200000,3650), 
                   np.random.normal(43000,100000,3650), 
                   np.random.normal(43500,140000,3650), 
                   np.random.normal(48000,70000,3650)], 
                  index=[1992,1993,1994,1995])
new_df = df.T
mean_values = new_df.mean()
std_deviation = new_df.std()
import numpy as np, scipy.stats as st

confidence_interval = st.t.interval(0.95, len(new_df)-1, loc=np.mean(new_df), scale=st.sem(new_df))
mar_err = []
col = 0
for row in range(confidence_interval[0].size):
    m = (confidence_interval[col+1][row] - confidence_interval[col][row])/2
    mar_err.append(m)

colors=["red", "white", "blue", "purple"]
fig = plt.figure()
ax = fig.add_subplot(111)

cm = plt.cm.get_cmap('RdYlBu')
plot = plt.scatter(range(len(new_df.columns)), mean_values, c=mean_values,cmap=cm)
#plt.clf()
plt.colorbar(plot)


ax.bar(range(len(new_df.columns)), mean_values, yerr=mar_err, align='center', alpha=0.5, color=colors,capsize=5,edgecolor='black')


def onclick(event): 
    
    plt.cla()
    
    colors = []
    col = 0
    for row in range(4):
        if ((event.ydata > confidence_interval[col+1][row] ) &  (event.ydata > confidence_interval[col][row] )):
            color_code = 'red'
        elif ((event.ydata < confidence_interval[col+1][row] ) &  ( event.ydata < confidence_interval[col][row])):
            color_code = 'blue'
        elif (( event.ydata < confidence_interval[col+1][row] ) & ( event.ydata > confidence_interval[col][row] )):
            color_code = 'white'
        else:
            color_code = ''
        colors.append(color_code)
    ax.bar(range(len(new_df.columns)), mean_values, yerr=mar_err, align='center', alpha=0.5, color=colors,capsize=5,edgecolor='black')
    plt.axhline(event.ydata,zorder=0)

cid = fig.canvas.mpl_connect('button_press_event', onclick) 


plt.xticks(range(len(new_df.columns)), new_df.columns)
plt.ylabel('Average Votes')
plt.title('Vote for President Elections.')


    
plt.show()
```
