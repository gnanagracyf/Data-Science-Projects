```python
 def convert(s):
    try:
        return float(s)
    except ValueError:
        num, denom = s.split(':')
        try:
            return float(num) / float(denom)
        except ZeroDivisionError:
            return float(num)
```

```python
convert("100:0")
```

```python
times_df['female_male_ratio'] = times_df['female_male_ratio'].map(convert)
```

```python
 def convert_world_ranking(s):
        try:
            if (s[0] == "="):
                s = s[1:]
            if (float(s) <= 10):
                return "1-10"
            elif (float(s) >= 11 and  float(s) <= 20 ):
                return "11-20"
            elif (float(s) >= 21 and  float(s) <= 30 ):
                return "21-30"
            elif (float(s) >= 31 and  float(s) <= 40 ):
                return "31-40"
            elif (float(s) >= 41 and  float(s) <= 50 ):
                return "41-50"
            elif (float(s) >= 51 and  float(s) <= 60 ):
                return "51-60"
            elif (float(s) >= 61 and  float(s) <= 70 ):
                return "61-70"
            elif (float(s) >= 71 and  float(s) <= 80 ):
                return "71-80"
            elif (float(s) >= 81 and  float(s) <= 90 ):
                return "81-90"
            elif (float(s) >= 91 and  float(s) <= 100 ):
                return "91-100"
            elif (float(s) >= 101 and  float(s) <= 150 ):
                return "101-150"
            elif (float(s) >= 151 and  float(s) <= 200 ):
                return "151-200"
        except ValueError:
            return s
```

```python
times_df['world_rank'] = times_df['world_rank'].map(convert_world_ranking)
```
