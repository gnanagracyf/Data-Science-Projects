This file contains additional data related to the store, department, and regional activity for the given dates. It contains the following fields:
Store - the store number Date - the week Temperature - average temperature in the region Fuel_Price - cost of fuel in the region MarkDown1-5 - anonymized data related to promotional markdowns that Walmart is running. MarkDown data is only available after Nov 2011, and is not available for all stores all the time. Any missing value is marked with an NA. CPI - the consumer price index Unemployment - the unemployment rate IsHoliday - whether the week is a special holiday week For convenience, the four holidays fall within the following weeks in the dataset (not all holidays are in the data):
Super Bowl: 12-Feb-10, 11-Feb-11, 10-Feb-12, 8-Feb-13 Labor Day: 10-Sep-10, 9-Sep-11, 7-Sep-12, 6-Sep-13 Thanksgiving: 26-Nov-10, 25-Nov-11, 23-Nov-12, 29-Nov-13 Christmas: 31-Dec-10, 30-Dec-11, 28-Dec-12, 27-Dec-13

```python
train_df = pd.read_csv('train.csv')
train_df.head()
```

```python
dataset = train_df.merge(stores, how = "left").merge(features, how="left")
```

```python
dataset.head()
```

```python
dataset_test = test_df.merge(stores, how = "left").merge(features, how="left")
```

```python
dataset_test['Store'] = dataset_test['Store'].astype(str)
dataset_test['Dept'] = dataset_test['Dept'].astype(str)
dataset_test['Store.Dept'] = dataset_test['Store'].map(str) + "." + dataset_test['Dept']
```

```python
X = dataset.drop(['Weekly_Sales','Dept','Store'], axis=1)
y = dataset['Weekly_Sales']
#Y_train = y_train.to_frame()
```

```python
import xgboost as xgb
data_dmatrix = xgb.DMatrix(data=X,label=y)
```

```python
model = xgb.XGBRegressor(objective ='reg:linear', colsample_bytree = 0.3, learning_rate = 0.1,max_depth = 5, alpha = 10, n_estimators = 10)
```

```python
model.fit(X,y)
```

```python
model.feature_importances_
```

```python
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, random_state=123)
```

```python
from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(y_val, predicts))
print("RMSE: %f" % (rmse))
```

```python
params = {"objective":"reg:linear",'colsample_bytree': 0.3,'learning_rate': 0.1,
                'max_depth': 5, 'alpha': 10}

cv_results = xgb.cv(dtrain=data_dmatrix, params=params, nfold=3,num_boost_round=50,early_stopping_rounds=10,metrics="rmse", as_pandas=True, seed=123)
```

```python
inport numpy as np
from sklearn.ensemble import ExtraTreesRegressor, RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from sklearn.feature_selection import RFE
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import KFold
```

```python
knn = KNeighborsRegressor(n_neighbors=10)
knn.fit(X_train,y_train)
pred = knn.predict(X_val)
error = np.sqrt(mean_squared_error(y_val,pred))
error
```

```python
clf = ExtraTreesRegressor(n_estimators=100,max_features='auto', verbose=1, n_jobs=1)
clf.fit(X_train,y_train)
pred = clf.predict(X_val)
error = np.sqrt(mean_squared_error(y_val,pred))
error
```

```python
clf = RandomForestRegressor(n_estimators=100,max_features='log2', verbose=1)
clf.fit(X_train,y_train)
pred = clf.predict(X_val)
error = np.sqrt(mean_squared_error(y_val,pred))
error
```

```python
clf = SVR(kernel='rbf', gamma='auto')
clf.fit(X_train,y_train)
pred = clf.predict(X_val)
error = np.sqrt(mean_squared_error(y_val,pred))
error
```

```python
clf = MLPRegressor(hidden_layer_sizes=(10,),  activation='relu', verbose=3)
clf.fit(X_train,y_train)
pred = clf.predict(X_val)
error = np.sqrt(mean_squared_error(y_val,pred))
error
```
