```python
# ingredients is a list...make it str with comma sepearated value. iterate thru each row.
data_train['ingredients'] = data_train.apply(lambda row: ','.join(row['ingredients']), axis=1)
```

```python
#find any special character like ( 0.5 oz.)  and remove it
import re
#str1 = "( 5 oz.)tomato sauce"
data_train['ingredients'] = data_train.apply(lambda row: re.sub("[\(\[].*?[\)\]]", "", (row['ingredients'])), axis=1)
```
