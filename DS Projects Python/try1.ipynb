```python
#Most Reviewed business
#Find the count of reviews for each business_id, juz take first 1000 and find category
review_count = review_df.groupby('business_id').count()[['review_id']].sort_values(by='review_id', ascending = False)
```

```python
review_count.rename(columns={'review_id':'rev_count'}, inplace=True)
review_count = review_count.iloc[0:1000]
```

```python
result_count = pd.merge(review_count, business_df, on='business_id')
```

```python
result_count = result_count[['business_id','name','rev_count','categories','latitude','longitude']]
```

```python
result_count
```

```python
result_count[0:20].plot(x="name",kind="bar",title="Top Reviewed Business").set_xlabel('Business Name')
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
from mpl_toolkits.basemap import Basemap
fig = plt.figure(figsize=(14, 8), edgecolor='w')

m = Basemap(projection='cyl',llcrnrlon= -180, urcrnrlon = 180,
            llcrnrlat = -90, urcrnrlat= 90,resolution='c',
           lat_ts = True)
m.drawcoastlines()
m.fillcontinents(color='#bbdaa4',lake_color='#FFFFFF')
m.drawcountries()
m.drawmapboundary(fill_color='#FFFFFF')

mloc = m(result_count['latitude'].tolist(),result_count['longitude'].tolist())
m.scatter(mloc[1],mloc[0],color ='red',lw=3,alpha=0.3,zorder=5)
```
